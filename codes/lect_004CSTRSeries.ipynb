The reaction A -> B, r = kCa occurs in n equal volume CSTRs in series, each with residence time tau, with 90% overall conversion. If k = 0.5 min^(-1), Cao = 2 mol/L, v = 4L/min, what residence times and reactor volumes will be required for 1, 2, 3 y 4?

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import pandas as pd
```

```python
params = {
    'axes.labelsize' : 'x-large',
    'ytick.labelsize' : 'large',
    'xtick.labelsize' : 'large'
}

rcParams.update(params)
plt.style.use('fivethirtyeight')
```

```python
x = 0.8        # adim      overall conversion
k = 0.5        # min^(-1)  rate constant.
Cao = 2        # mol/L     inicial concentration
qo = 4          # L/min     flow rate
nreactor = 8   # Reactor numbers
```

```python
nreactors = np.arange(1, nreactor + 1)
taus = nreactors/qo

conversions = [1 - 1/(1 + tau * k)**nreactors for tau in taus]    
```

Next table shows the reached conversion as function of $\tau$ and the number of reactors in the system

```python
tableConversion = pd.DataFrame(conversions, columns=nreactors, index=taus)
```

```python
tableConversion
```

```python
plt.figure(figsize=(7, 7))
plt.plot(nreactors, tableConversion, marker='o',linestyle= '-')

plt.annotate(text='k$\\tau$=' + str(k*taus[-1]), xy=(2.6, 0.9), xytext=(1.5, 1), arrowprops=dict(facecolor='black'))
plt.annotate(text='k$\\tau$=' + str(k*taus[0]), xy=(3.8, 0.3), xytext=(4, 0.2), arrowprops=dict(facecolor='black'))
plt.xlabel('Number of CSTRs')
plt.ylabel('Conversion, $x_A$')
plt.show()
```
